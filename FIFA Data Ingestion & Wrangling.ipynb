Here we write a function that shows us all players from a specific football club, to maybe look into some interesting similarities or other. 

But the player_id = 179897 seems odd. We can see, that the player has a questionmark inside his name. This must be because of some character that is unreadable inside our dataset, so let's see if we can do anything about it.

```python
def players_from_club(df, club_name):
    return df[df['team'].str.contains(club_name, case=False, na=False)]
```

```python
players_from_club(df, "FC København")
```

Same function as before, just for nationalities instead of football clubs.

```python
def players_from_nation(df, nation_name):
    return df[df['nationality'].str.contains(nation_name, case=False, na=False)]
```

```python
players_from_nation(df, "Iceland").head(20)
```

Here we write a function, that can search for different charaters in the name column. We do this to check for all players with an unrecognized character inside their name.

We see that there are 95 players with an unrecognized character, and sadly after searching up some of the names, there are multiple different unrecognizable characters, so we cant just replace them all with the same character.

We looked into our dataset, to check and see if maybe the characters could be restored, but sadly the dataset have also marked them as questionmarks.

We also see for player_id = 16999 his football club also has an unrecognizable character, but it's the same story as for the players.

```python
def players_with_questionmark(df, char):
    return df[df['name'].str.contains(char, case=False, na=False, regex=False)]
```

```python
filtered_df = players_with_questionmark(df, "?")
```
